### Step 2.5 — Apply Randomized Discount Based on Roll Number

- **Even roll number** → Apply **10% discount** to the **Electronics** category
- **Odd roll number** → Apply **15% discount** to the **Clothing** category

````python
# ── Step 2.5: Apply randomized discount based on roll number ───────────────
print(f'Roll Number: {ROLL_NUMBER}  →  {"EVEN" if ROLL_NUMBER % 2 == 0 else "ODD"}')

if ROLL_NUMBER % 2 == 0:
    # Even roll: 10% discount on Electronics
    discount_col  = 'Electronics'
    discount_rate = 0.10
    print(f'Applying 10% discount to {discount_col}...')
else:
    # Odd roll: 15% discount on Clothing
    discount_col  = 'Clothing'
    discount_rate = 0.15
    print(f'Applying 15% discount to {discount_col}...')

# Store original for comparison
original_col = df[discount_col].copy()

# Apply discount: reduce by discount_rate
df[discount_col] = (df[discount_col] * (1 - discount_rate)).round(0).astype(int)

# Also update Total Sales since category values changed
df['Total Sales'] = df[categories].sum(axis=1)
df['Growth Rate (%)'] = df['Total Sales'].pct_change() * 100

# Show comparison
comparison = pd.DataFrame({
    f'{discount_col} (Original)': original_col,
    f'{discount_col} (After {int(discount_rate*100)}% Discount)': df[discount_col],
    'Savings (units)': original_col - df[discount_col]
})
print()
print(comparison)
print(f'\nTotal discount applied: {(original_col - df[discount_col]).sum():,} units')
````

````python
# ── Q2 (i): Element-wise Absolute Value ───────────────────────────────────
print('═' * 50)
print('Q2 (i) — Element-wise Absolute Value')
print('═' * 50)
print('Original array:')
print(array)
print()
abs_array = np.abs(array)
print('np.abs(array):')
print(abs_array)
print()
print('Element-wise comparison:')
for r in range(array.shape[0]):
    for c in range(array.shape[1]):
        print(f'  array[{r},{c}] = {array[r,c]:3d}  →  |{array[r,c]:3d}| = {abs_array[r,c]}')
````

````python
# ── Q2 (ii): Percentiles ──────────────────────────────────────────────────
print('═' * 55)
print('Q2 (ii) — 25th, 50th, 75th Percentiles')
print('═' * 55)

percentiles = [25, 50, 75]
labels      = ['25th (Q1)', '50th (Q2/Median)', '75th (Q3)']

# (a) Flattened array
print('\n(a) Percentiles of FLATTENED array:')
flat = array.flatten()
print(f'    Flattened: {flat}')
for p, lbl in zip(percentiles, labels):
    val = np.percentile(flat, p)
    print(f'    {lbl}: {val}')

# (b) Per column (axis=0)
print('\n(b) Percentiles for EACH COLUMN (axis=0):')
print('    Columns → Col 0, Col 1, Col 2')
for p, lbl in zip(percentiles, labels):
    val = np.percentile(array, p, axis=0)
    print(f'    {lbl}: {val}')

# (c) Per row (axis=1)
print('\n(c) Percentiles for EACH ROW (axis=1):')
print('    Rows → Row 0, Row 1')
for p, lbl in zip(percentiles, labels):
    val = np.percentile(array, p, axis=1)
    print(f'    {lbl}: {val}')
````

````python
# ── Q2 (iii): Mean, Median, Standard Deviation ────────────────────────────
print('═' * 55)
print('Q2 (iii) — Mean, Median, Standard Deviation')
print('═' * 55)

stats = {
    'Mean'  : np.mean,
    'Median': np.median,
    'Std Dev': np.std,
}

# (a) Flattened
print('\n(a) Statistics of FLATTENED array:')
flat = array.flatten()
print(f'    Flattened: {flat}')
for stat_name, func in stats.items():
    print(f'    {stat_name:<10}: {func(flat):.4f}')

# (b) Per column
print('\n(b) Statistics per COLUMN (axis=0):')
print(f'    Columns     : Col 0  Col 1  Col 2')
for stat_name, func in stats.items():
    vals = func(array, axis=0)
    print(f'    {stat_name:<10}: {vals}')

# (c) Per row
print('\n(c) Statistics per ROW (axis=1):')
print(f'    Rows        : Row 0   Row 1')
for stat_name, func in stats.items():
    vals = func(array, axis=1)
    print(f'    {stat_name:<10}: {vals}')
````

# Q3 — NumPy Rounding Functions

**Given array:**
```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
```
Find: **floor**, **ceiling**, **truncated**, and **rounded** values.

````python
# ── Q3: Floor, Ceiling, Truncation, Rounding ──────────────────────────────
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print('═' * 70)
print('Q3 — NumPy Rounding Functions')
print('═' * 70)
print(f'Original array: {a}')
print()

# Compute all four
floor_vals = np.floor(a)      # largest integer ≤ x
ceil_vals  = np.ceil(a)       # smallest integer ≥ x
trunc_vals = np.trunc(a)      # removes decimal part (toward zero)
round_vals = np.round(a)      # rounds to nearest integer (half to even)

print(f'{"Value":<10} {"Floor":<10} {"Ceiling":<10} {"Truncate":<12} {"Round":<10}')
print('-' * 60)
for orig, fl, ce, tr, ro in zip(a, floor_vals, ceil_vals, trunc_vals, round_vals):
    print(f'{orig:<10.1f} {fl:<10.1f} {ce:<10.1f} {tr:<12.1f} {ro:<10.1f}')

print()
print('Definitions:')
print('  np.floor(x)  → Largest integer ≤ x           (rounds toward -∞)')
print('  np.ceil(x)   → Smallest integer ≥ x           (rounds toward +∞)')
print('  np.trunc(x)  → Drops the decimal part          (rounds toward 0)')
print('  np.round(x)  → Nearest integer (ties → even)   (banker rounding)')
print()
print('Key difference — trunc vs floor for negatives:')
print('  np.floor(-1.6) =', np.floor(-1.6), '  (goes further negative)')
print('  np.trunc(-1.6) =', np.trunc(-1.6), '  (goes toward zero)')
````

# Q4 — Swap Two Elements in a List Using a Temporary Variable

**Objective:** Swap two specific elements in a Python list using the classical **temporary variable** technique.

````python
# ── Q4: Swap Two Elements in a List Using a Temporary Variable ─────────────
print('═' * 55)
print('Q4 — Swap Two List Elements (Temporary Variable)')
print('═' * 55)

def swap_list_elements(lst, idx1, idx2):
    """
    Swap elements at idx1 and idx2 in lst using a temporary variable.

    Parameters:
        lst  : list — the input list
        idx1 : int  — index of first element to swap
        idx2 : int  — index of second element to swap

    Returns:
        list with the two elements swapped
    """
    # Validate indices
    if idx1 < 0 or idx2 < 0 or idx1 >= len(lst) or idx2 >= len(lst):
        raise IndexError(f'Index out of range for list of length {len(lst)}')

    # Step 1: Store element at idx1 in a temporary variable
    temp = lst[idx1]

    # Step 2: Assign element at idx2 to position idx1
    lst[idx1] = lst[idx2]

    # Step 3: Assign the saved temp value to position idx2
    lst[idx2] = temp

    return lst

# ── Example 1: Integer list ────────────────────────────────────────────────
lst1 = [10, 20, 30, 40, 50]
idx1, idx2 = 1, 4
print(f'\nExample 1 — Integer list')
print(f'  Original list    : {lst1}')
print(f'  Swap indices      : {idx1} and {idx2}')
print(f'  Element at [{idx1}]   : {lst1[idx1]}')
print(f'  Element at [{idx2}]   : {lst1[idx2]}')
swap_list_elements(lst1, idx1, idx2)
print(f'  After swap        : {lst1}')

# ── Example 2: String list ─────────────────────────────────────────────────
lst2 = ['apple', 'banana', 'cherry', 'date', 'elderberry']
idx1, idx2 = 0, 3
print(f'\nExample 2 — String list')
print(f'  Original list    : {lst2}')
print(f'  Swap indices      : {idx1} and {idx2}')
swap_list_elements(lst2, idx1, idx2)
print(f'  After swap        : {lst2}')

# ── Trace the steps explicitly ─────────────────────────────────────────────
print(f'\n── Step-by-step Trace (for Example 1) ──')
lst_trace = [10, 20, 30, 40, 50]
i, j = 1, 4
print(f'  list before   : {lst_trace}')
print(f'  Step 1: temp  = lst[{i}] = {lst_trace[i]}')
temp = lst_trace[i]
lst_trace[i] = lst_trace[j]
print(f'  Step 2: lst[{i}] = lst[{j}]  → list = {lst_trace}')
lst_trace[j] = temp
print(f'  Step 3: lst[{j}] = temp({temp}) → list = {lst_trace}')
````

# Q5 — Swap Two Elements in a Set by Converting to List

**Objective:** Python sets are **unordered** and do not support indexing. To swap two elements, we must:
1. Convert the set to a list
2. Perform the swap on the list
3. Convert back to a set (or keep as list with swapped values)

> **Note:** Since sets are unordered, the concept of "position" doesn't apply directly. We swap by **value** — removing one element and adding the other in its place.

````python
# ── Q5: Swap Two Elements in a Set by Converting to List ──────────────────
print('═' * 60)
print('Q5 — Swap Two Set Elements (Convert to List)')
print('═' * 60)

def swap_set_elements(s, val1, val2):
    """
    Swap two elements in a set by converting to a list.

    Since sets are unordered, 'swapping' means:
      → Remove val1, add val2 (and vice versa) effectively
        replacing val1 with val2 and val2 with val1.
    In a list representation, find their indices and swap.

    Parameters:
        s    : set  — the input set
        val1 : element to swap
        val2 : element to swap

    Returns:
        tuple: (updated list, updated set)
    """
    if val1 not in s:
        raise ValueError(f'{val1} not found in set')
    if val2 not in s:
        raise ValueError(f'{val2} not found in set')

    # Step 1: Convert set to list
    lst = list(s)
    print(f'  Step 1 — Convert set to list    : {lst}')

    # Step 2: Find indices of the two values
    idx1 = lst.index(val1)
    idx2 = lst.index(val2)
    print(f'  Step 2 — Index of {val1!r}: {idx1}, Index of {val2!r}: {idx2}')

    # Step 3: Swap using temporary variable
    temp     = lst[idx1]
    lst[idx1] = lst[idx2]
    lst[idx2] = temp
    print(f'  Step 3 — After swap in list     : {lst}')

    # Step 4: Convert back to set
    new_set = set(lst)
    print(f'  Step 4 — Convert list back to set: {new_set}')
    print(f'  (Note: set order may differ — sets are unordered by design)')

    return lst, new_set

# ── Example 1: Integer set ─────────────────────────────────────────────────
my_set = {10, 20, 30, 40, 50}
elem1, elem2 = 20, 40

print(f'\nExample 1 — Integer set')
print(f'  Original set     : {my_set}')
print(f'  Swapping values  : {elem1} ↔ {elem2}')
result_list, result_set = swap_set_elements(my_set, elem1, elem2)
print(f'  Resulting list   : {result_list}')
print(f'  Resulting set    : {result_set}')

# ── Example 2: String set ──────────────────────────────────────────────────
str_set = {'alpha', 'beta', 'gamma', 'delta'}
elem1, elem2 = 'alpha', 'gamma'

print(f'\nExample 2 — String set')
print(f'  Original set     : {str_set}')
print(f'  Swapping values  : {elem1!r} ↔ {elem2!r}')
result_list, result_set = swap_set_elements(str_set, elem1, elem2)
print(f'  Resulting list   : {result_list}')
print(f'  Resulting set    : {result_set}')

# ── Key Observations ───────────────────────────────────────────────────────
print()
print('Key Observations:')
print('  • Sets are unordered → no inherent position/index')
print('  • Conversion to list gives a positional structure to work with')
print('  • After swapping in the list, converting back to set gives the same elements')
print('  • The list representation is more meaningful when swap order matters')

print()
print('=' * 60)
print('Lab Assignment 3 — ALL QUESTIONS COMPLETE ✅')
print('=' * 60)
````
