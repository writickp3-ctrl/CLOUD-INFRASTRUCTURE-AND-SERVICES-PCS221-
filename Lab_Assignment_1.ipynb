## Task 1: Introduction to Google Colab Cloud

**Google Colaboratory (Colab)** is a free, cloud-based Jupyter Notebook environment provided by Google. It allows you to write and execute Python code directly in your browser without any local setup.

### Key Features:
- **Free GPU/TPU access** for machine learning and deep learning tasks
- **No installation required** — runs entirely in the browser
- **Integration with Google Drive** for saving and loading files
- **Collaborative editing** — multiple users can work simultaneously
- **Pre-installed libraries** — NumPy, Pandas, TensorFlow, PyTorch, etc.
- **Persistent storage** via Google Drive mounting

### Why Colab for Cloud Computing?
Colab is essentially a managed cloud computing environment. When you run code, it executes on Google's remote servers — not your local machine. This makes it ideal for:
- Running compute-intensive programs without expensive hardware
- Sharing reproducible research and code
- Learning distributed and cloud-based computation

```python
# Task 1: Verify we are running in a cloud environment
import platform
import sys
import os

print("=" * 50)
print("GOOGLE COLAB CLOUD ENVIRONMENT INFO")
print("=" * 50)
print(f"Operating System  : {platform.system()} {platform.release()}")
print(f"Python Version    : {sys.version}")
print(f"Machine Type      : {platform.machine()}")
print(f"Processor         : {platform.processor()}")
print(f"Current Directory : {os.getcwd()}")
print()

# Check if running in Colab
try:
    import google.colab
    print("✅ Running inside Google Colab Cloud Environment")
except ImportError:
    print("ℹ️  Not running in Colab (running locally)")
```

## Task 2: Interacting with the Colab Environment

Google Colab provides several ways to interact with the cloud environment:

| Interaction | Method |
|-------------|--------|
| Run shell commands | Use `!` prefix (e.g., `!ls`, `!pwd`) |
| Install packages | `!pip install <package>` |
| Mount Google Drive | `from google.colab import drive` |
| Upload local files | `from google.colab import files` |
| Use magic commands | `%timeit`, `%matplotlib inline`, etc. |

```python
# Task 2: Interacting with the Colab environment

# 2a: Shell commands using ! prefix
print("--- Current Directory Contents ---")
!ls

print("\n--- Disk Usage ---")
!df -h

print("\n--- Memory Info ---")
!cat /proc/meminfo | head -5
```

```python
# 2b: Using Python magic commands
import time

# %timeit measures execution time
print("Timing a simple computation using %timeit:")
%timeit sum(range(10000))

# Environment variables
print("\nSome Environment Variables:")
for key in ['HOME', 'PATH', 'USER']:
    print(f"  {key} = {os.environ.get(key, 'Not Set')}")
```

```python
# 2c: Writing and reading files in the Colab environment
# Writing a sample file
with open('sample_colab_file.txt', 'w') as f:
    f.write("Hello from Google Colab!\n")
    f.write("This file is stored on the cloud VM.\n")
    f.write("It will be lost when the runtime disconnects.\n")

# Reading it back
with open('sample_colab_file.txt', 'r') as f:
    content = f.read()

print("File written and read successfully:")
print(content)
print("File size:", os.path.getsize('sample_colab_file.txt'), "bytes")
```

## Task 3: Google Colab Computational Resources

Colab provides access to the following computational resources:

### CPU
- Standard Intel Xeon processors
- Multiple cores available for parallel tasks

### GPU (Graphics Processing Unit)
- NVIDIA Tesla K80, T4, or P100 (varies by availability)
- Essential for deep learning model training
- Access via: Runtime → Change Runtime Type → GPU

### TPU (Tensor Processing Unit)
- Google's custom AI accelerator chip
- Fastest option for TensorFlow workloads

### RAM & Storage
- ~12 GB RAM (standard), ~25 GB (Pro)
- ~100 GB disk space on the VM
- Unlimited storage via Google Drive

```python
# Task 3: Check computational resources available
import psutil

print("=" * 50)
print("COMPUTATIONAL RESOURCES IN COLAB")
print("=" * 50)

# CPU Info
print(f"\n📦 CPU Cores (Logical) : {psutil.cpu_count(logical=True)}")
print(f"📦 CPU Cores (Physical): {psutil.cpu_count(logical=False)}")
print(f"📦 CPU Usage           : {psutil.cpu_percent(interval=1)}%")

# RAM Info
ram = psutil.virtual_memory()
print(f"\n💾 Total RAM  : {ram.total / (1024**3):.2f} GB")
print(f"💾 Used RAM   : {ram.used  / (1024**3):.2f} GB")
print(f"💾 Free RAM   : {ram.available / (1024**3):.2f} GB")
print(f"💾 RAM Usage  : {ram.percent}%")

# Disk Info
disk = psutil.disk_usage('/')
print(f"\n💿 Total Disk : {disk.total / (1024**3):.2f} GB")
print(f"💿 Used Disk  : {disk.used  / (1024**3):.2f} GB")
print(f"💿 Free Disk  : {disk.free  / (1024**3):.2f} GB")
```

```python
# Check GPU availability
print("\n🎮 GPU Information:")
try:
    !nvidia-smi
except:
    print("No GPU detected (switch Runtime to GPU for GPU access)")

# Benchmark: measure computation speed
import numpy as np
import time

print("\n⚡ Compute Benchmark — Matrix Multiplication (1000x1000):")
A = np.random.rand(1000, 1000)
B = np.random.rand(1000, 1000)
start = time.time()
C = np.dot(A, B)
elapsed = time.time() - start
print(f"   Time taken : {elapsed:.4f} seconds")
print(f"   Result shape: {C.shape}")
```

## Task 4: Python Environment in Google Colab Cloud

Colab comes with a rich Python environment pre-loaded with hundreds of scientific and ML libraries. Below we explore the environment and demonstrate key capabilities.

```python
# Task 4: Explore the Python environment
import pkg_resources

# Key pre-installed libraries in Colab
key_packages = [
    'numpy', 'pandas', 'matplotlib', 'scipy',
    'sklearn', 'tensorflow', 'torch', 'PIL'
]

print("=" * 50)
print("PRE-INSTALLED KEY PYTHON LIBRARIES IN COLAB")
print("=" * 50)

for pkg in key_packages:
    try:
        mod = __import__(pkg)
        ver = getattr(mod, '__version__', 'N/A')
        print(f"  ✅ {pkg:<15} version: {ver}")
    except ImportError:
        print(f"  ❌ {pkg:<15} not installed")
```

```python
# 4b: Demonstrate Python capabilities in Colab — Plotting
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 2 * np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(x, y1, label='sin(x)', color='blue', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linewidth=2)
plt.title('Python Plotting in Google Colab Cloud')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
print("Plot rendered successfully inside Colab cloud environment!")
```
